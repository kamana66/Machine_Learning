## Numpy

**NumPy (Numerical Python) provides:**

-The ndarray — a fast, memory-efficient N-dimensional array object.

-Vectorized operations (C-backed) via ufuncs for elementwise computations.

-Broadcasting rules to operate on arrays of different shapes.

-Tools for linear algebra, random number generation, FFTs, and I/O.

- Numpy stands for numerical python
- Building block for many other libraries like Pandas, Matplotlib etc

 **ndarray**

Homogeneous: every element has the same dtype (e.g., float64, int32).

**Attributes:**

-ndarray.shape — tuple for dimensions (rows, cols, ...).

-ndarray.ndim — number of dimensions.

-ndarray.size — total number of elements.

-ndarray.dtype — data type of elements.

-ndarray.itemsize — bytes per element.

-ndarray.nbytes — total bytes (size * itemsize).

-ndarray.strides — bytes to step in each dimension.

### Built in arrays:
- arange - a 1-d array with uniformly spaced numbers between a start and stop value. Integers
  - np.arange(start,stop,step)
  - Can be used in combination with a reshape function to create an array of any dimension. 
- zeros, ones - zero  matrix and ones matrix
  - np.zeros(shape tuple,dtype)
  - np.ones(shape tuple,dtype)
- linspace - Generate a count of samples between start and stop values
  - np.linspace(start,stop,count)
  - Stop is inclusive
- eye - Create an identity matrix
  - (defaults square matrix, can be made as rectangular matrix)

### Random module
#### Functions 
- rand: Generate a float array of chosen dimension of values between 0 and 1
- randn: Generates a array of chosen dimension with values that belong a Standard normal distribution
- randint: Generates a array of integers between chosen values

```python
# rand
#from numpy import random
np.random.rand(3,4)   # 0 is inclusive, 1 is exclusive
```

```python
np.random.randn(3,4) # -3 to +3
```

```python
#np.random.randint(low,high,count)
np.random.randint(2,50,12)
```

```python
np.random.randint(1,7,40)  #Generate 40 random integers between 1 and 7. End point is exclusive
```

```python
np.random.randint(2,10,(2,5))     #Generate a array of shape (2,5) with random integers from 2 to 10, 10 is exclusive
```
